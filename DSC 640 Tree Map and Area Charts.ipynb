```python
##View os and imports
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import os
import squarify
print(os)
```

```python
## Import Excel
df = pd.read_csv('unemployement-rate-1948-2010.csv')
```

```python
## View Data
df.head(10)
```

```python
##Create tree map to show unemployment by year
fig, ax = plt.subplots(1, figsize = (12,12))
squarify.plot(sizes=df['Value'], 
              label=df['Year'], 
              alpha=.8 )
plt.axis('off')
plt.show()
```

```python
##Create area chart to show unemployment by year
x = df['Year']
y = df['Value']
 
##Setting up chart params
plt.fill_between( x, y, color='skyblue', alpha=0.2)
plt.title('Unemployment by Year')
plt.xlabel('Year')
plt.ylabel('Unemployment')
plt.plot(x, y, color='darkblue', alpha=0.6)
```
