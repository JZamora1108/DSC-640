```python
##Import Excel for different option
df = pd.read_csv('birth-rates-yearly.csv')
```

```python
df.head(10)
```

```python
##Set up scatterplot to show birth rates by year
plt.style.use('fivethirtyeight')

x = df['year']
y = df['rate']

plt.scatter(x, y)
plt.xlabel('Year')
plt.ylabel('Rate')
plt.show()
```

```python
##Set up bubble chart 

x = df['year']
y = df['rate']

plt.scatter(x, y, s = x, alpha = 0.5, c = x)
plt.xlabel('Year')
plt.ylabel('Rate')
plt.show()
```

```python
##Set up density plot

df.year.plot.density(color='blue')
plt.title('Density Plot by Year')
plt.show()
```
